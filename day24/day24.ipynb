```python
from collections import namedtuple
import itertools
import re
```

```python
OpcodeBase = namedtuple("OpcodeBase", ["code", "a", "b"])

class Opcode(OpcodeBase):
    tostr = {
        "inp": lambda s: f"inp {s.a}",
        "add": lambda s: f"{s.a} += {s.b}",
        "mul": lambda s: f"{s.a} *= {s.b}",
        "div": lambda s: f"{s.a} //= {s.b}",
        "mod": lambda s: f"{s.a} %= {s.b}",
        "eql": lambda s: f"{s.a} = ({s.a} == {s.b})",
        "set": lambda s: f"{s.a} = {s.b}"
    }
    
    def __str__(self):
        return Opcode.tostr[self.code](self)
    
patt_inp = re.compile(r"^inp (.)$")
patt_bin = re.compile(r"^(...) (.) (.+)$")
patt_var = re.compile(r"^[wxyz]$")
patt_int = re.compile(r"-?[0-9]+$")

def parse(s):
    m = patt_inp.match(s)
    if m:
        a = m.group(1)
        if patt_var.match(a):
            return Opcode("inp", a, None)
    m = pattbin.match(s)
    if m:
        code = m.group(1)
        a = m.group(2)
        b = m.group(3)
        if patt_var.match(a) and patt_int.match(b):
            return Opcode(code, a, int(b))
        elif patt_var.match(a) and patt_var.match(b):
            return Opcode(code, a, b)
    return None

inp = []
with open("input.txt", "r") as fh:
    for l in fh:
        l = l.strip()
        op = parse(l)
        inp.append(op)
        
for i in inp:
    print(i)
```

```python
def simplify(program):
    gen = (op for op in program)
    out = []
    try:
        while True:
            op1 = next(gen)
            if op1.code == "div" and op1.b == 1:
                continue
            elif op1.code == "add" and op1.b == 0:
                continue
            elif op1.code == "mul" and op1.b == 0:
                op2 = next(gen)
                if op2.code == "add" and op1.a == op2.a:
                    out.append(Opcode("set", op1.a, op2.b))
                else:
                    out.append(op1)
                    out.append(op2)
            else:
                out.append(op1)
    except:
        pass
    return out

for op in simplify(inp):
    print(op)
```

```python
def evaluate(program, inpdata, stop=20):
    nimp = 0
    inp = (x for x in inpdata)
    reg = {"w": 0, "x": 0, "y": 0, "z": 0}
    
    def value(b):
        if isinstance(b, int):
            return b
        else:
            return reg[b]
    
    for op in program:
        if op.code == "inp":
            reg[op.a] = next(inp)
            nimp += 1
            if nimp >= stop:
                break
        elif op.code == "add":
            reg[op.a] += value(op.b)
        elif op.code == "mul":
            reg[op.a] *= value(op.b)
        elif op.code == "mod":
            reg[op.a] %= value(op.b)
        elif op.code == "div":
            reg[op.a] //= value(op.b)
        elif op.code == "eql":
            if reg[op.a] == value(op.b):
                reg[op.a] = 1
            else:
                reg[op.a] = 0
    return reg

evaluate(inp, [1, 2, 3, 4, 5, 6, 7, 8, 9, 1, 2, 3, 4, 5])
```

```python
for x in range(1, 10):
    for y in range(1, 10):
        out = evaluate(inp, inpdata=[x, y], stop=2)
        print(x, y, out)
```
