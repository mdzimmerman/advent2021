```python
import itertools
```

```python
class Die:
    def __init__(self, sides=100):
        self.sides = sides
        self.i = 0

    def __iter__(self):
        return self

    def __next__(self):
        self.i += 1
        if self.i > self.sides:
            self.i = 1
        return self.i
        
die = Die(100)
```

```python
list(itertools.islice(die, 3))
```

```python
def toggle(turn):
    if turn == 1:
        return 0
    else:
        return 1

def play(start):
    pos = start.copy()
    scores = [0, 0]
    turn = 0
    nroll = 0
    die = Die(100)
    while max(scores) < 1000:
        roll = list(itertools.islice(die, 3))
        nroll += 3
        npos = (pos[turn] + sum(roll)) % 10
        if npos == 0:
            pos[turn] = 10
        else:
            pos[turn] = npos
        scores[turn] += pos[turn]
        print(f"turn={turn} roll={roll} pos={pos} scores={scores} nroll={nroll}")
        turn = toggle(turn)
    return min(scores) * nroll

play([4, 8])
```
